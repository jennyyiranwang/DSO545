```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
## Question 1

plt.figure()

## existed density plot function 
sns.kdeplot(mpg.loc[mpg.cyl == 4,'cty'],
            shade = True,
            color = 'green',
            label = "Cyl = 4",
            alpha = 0.4
            ) 

sns.kdeplot(mpg.loc[mpg.cyl == 5,'cty'],
            shade = True,
            color = 'orchid',
            label = "Cyl = 5",
            alpha = 0.4
            ) 

sns.kdeplot(mpg.loc[mpg.cyl == 6,'cty'],
            shade = True,
            color = 'cornflowerblue',
            label = "Cyl = 6",
            alpha = 0.4
            ) 

sns.kdeplot(mpg.loc[mpg.cyl ==8,'cty'],
            shade = True,
            color = 'cornsilk',
            label = "Cyl = 8",
            alpha = 0.4
            ) 
plt.xlim(5,35)
plt.ylim(0,0.5)

plt.legend()
plt.show()
```

```python
plt.figure()
sns.kdeplot(mpg.loc[mpg.cyl == 5,'hwy'],
            shade = True,
            color = 'powderblue',
            label = "Cyl = 5",
            alpha = 0.4
            ) 
sns.kdeplot(mpg.loc[mpg.cyl == 5,'cty'],
            shade = True,
            color = 'orchid',
            label = "Cyl = 5",
            alpha = 0.4
            ) 
plt.show()
```

```python
## Question 3 

plt.figure()

## sharex and sharey means that they all share one x and y axis 
fig,ax = plt.subplots(2,3,sharex = 'col', sharey = 'row')

## create the dataset
## linspace = linearspace which is like nrange 
x = np.linspace(0,50)
ax[0,0].plot(np.sin(x)) ##looking at first square and assign it to sin
ax[1,2].plot(np.cos(x), color = 'red')

plt.show()
```

```python
## Question 4

fig = plt.figure()

grid = plt.GridSpec(2,3, wspace = 0.4, hspace = 0.5)

fig.add_subplot(grid[0,0])
fig.add_subplot(grid[0,1:3])
fig.add_subplot(grid[1,0:2])
fig.add_subplot(grid[1,2])


plt.show()
```

```python
## Question 5: Marginal Boxplots 

fig = plt.figure()

## set the grids
grid = plt.GridSpec(4,4, wspace = 0.4, hspace = 0.5)

ax_main = fig.add_subplot(grid[0:3, 0:3])
ax_right = fig.add_subplot(grid[0:3,3])
ax_bottom = fig.add_subplot(grid[3,0:3])

## plot the main graph
ax_main.scatter(mpg.displ, mpg.hwy)
ax_main.set(xticks = np.arange(2,7,2), yticks = np.arange(20,50,10))

## plot the right boxplot
sns.boxplot(mpg.hwy, ax = ax_right, orient = 'v')
ax_right.set(xticks = [], yticks = [], ylabel = '')

## plot bottom
sns.boxplot(mpg.displ, ax = ax_bottom, orient = 'h')
ax_bottom.set(xticks = [], yticks = [], xlabel = '')

ax_main.set(title = "Scatterplot with Boxplots \n Displacement vs Highway Mileage",
            xlabel = 'Displacement',
            ylabel = 'hwy')

plt.show()
```

```python
## Question 6 

econ = pd.read_csv("economics.csv")
econ.head()
```

```python
fig = plt.figure(figsize = (12,5))

ax = plt.gca()

plt.plot(econ.date, econ.psavert, color = 'tab:red')

## set the main yaxis
ax.tick_params(axis = 'y', labelcolor = 'tab:red')
ax.set_ylabel("Personal Savings Rate", color = 'tab:red', fontsize = 14)

## set the secondary yaxis
ax_right = ax.twinx() ## create a twin yaxis to share the same xaxis
ax_right.plot(econ.date, econ.unemploy, color = 'tab:blue')
ax_right.tick_params(axis = 'y', labelcolor = 'tab:blue')
ax_right.set_ylabel("Unemployment (1000's)", color = 'tab:blue', fontsize = 14)


## set the x axis so it doesn't take so long to load 
ax.set_xticks(np.arange(0,len(econ.date),60))

plt.title("Personal Savings Rate vs Unemployment", fontsize = 20)

plt.show()
```
